```python
import pandas as pd
import glob
import os
from os.path import abspath
```

```python
'''Retrieve and read last stored csv file'''
data_path = abspath('./data/')
lst = os.listdir(data_path)
lst.sort(reverse=True)

dir = lst[0]
dir_path = data_path + '/' + dir 
file_path = glob.glob(dir_path + '/*')[0]

df = pd.read_csv(file_path)
df.sort_values(by='ranking')
```
